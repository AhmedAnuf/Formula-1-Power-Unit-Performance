```python
import numpy as np
import pandas as pd
from numpy.linalg import solve
```

```python
df = pd.read_csv('football_matches_2024_2025.csv')
```

```python
df.head
```

```python
ucl_all = df[
    (df["competition_name"] == "UEFA Champions League") &
    (df["season"] == "2024/2025")
]

print(ucl_all)
```

```python
ucl_league= df[
    (df["competition_name"] == "UEFA Champions League") &
    (df["season"] == "2024/2025") &
    (df["stage"] == "LEAGUE_STAGE")
]


print(ucl_league)
```

```python
# Home team perspective
home = ucl_league[[
    "match_id",
    "home_team",
    "away_team",
    "fulltime_home",
    "fulltime_away"
]].copy()

home["team"] = home["home_team"]
home["opponent"] = home["away_team"]

home["result"] = home.apply(
    lambda x: "Win" if x["fulltime_home"] > x["fulltime_away"]
    else "Draw" if x["fulltime_home"] == x["fulltime_away"]
    else "Loss",
    axis=1
)

home = home[["match_id", "team", "opponent", "result"]]
```

```python
# Away team perspective
away = ucl_league[[
    "match_id",
    "home_team",
    "away_team",
    "fulltime_home",
    "fulltime_away"
]].copy()

away["team"] = away["away_team"]
away["opponent"] = away["home_team"]

away["result"] = away.apply(
    lambda x: "Win" if x["fulltime_away"] > x["fulltime_home"]
    else "Draw" if x["fulltime_away"] == x["fulltime_home"]
    else "Loss",
    axis=1
)

away = away[["match_id", "team", "opponent", "result"]]
```

```python
team_match_results = pd.concat([home, away], ignore_index=True)
```

```python
team_name1 = "Arsenal FC"
team_name2 = "AC Milan"

arsenal_matches = team_match_results[
    team_match_results["team"] == team_name1
]

print(arsenal_matches)

acMilan_matches = team_match_results[
    team_match_results["team"] == team_name2
]

print(acMilan_matches)
```

```python
formatted_output = arsenal_matches.apply(
    lambda x: f"{x['team']} | {x['opponent']} | {x['result']}",
    axis=1
)

print(formatted_output.to_string(index=False))
```

```python
#COLLEY RANK
```

```python
df = team_match_results.copy()

df["team"] = df["team"].astype(str).str.strip()
df["opponent"] = df["opponent"].astype(str).str.strip()
df["result"] = df["result"].astype(str).str.strip().str.lower()
```

```python
matches = (
    df.sort_values(["match_id", "team"])
      .groupby("match_id")
      .agg({"team": list, "result": list})
      .reset_index()
)

matches["team_i"] = matches["team"].apply(lambda x: x[0])
matches["team_j"] = matches["team"].apply(lambda x: x[1])
matches["result_i"] = matches["result"].apply(lambda x: x[0])
matches["result_j"] = matches["result"].apply(lambda x: x[1])

matches = matches[["match_id", "team_i", "team_j", "result_i", "result_j"]]

print("matches:", len(matches))
print(matches.head())
```

```python
bad = matches[
    ~(
        ((matches["result_i"] == "win") & (matches["result_j"] == "loss")) |
        ((matches["result_i"] == "loss") & (matches["result_j"] == "win")) |
        ((matches["result_i"] == "draw") & (matches["result_j"] == "draw"))
    )
]

print("bad matches:", len(bad))
bad.head()
```

```python
teams = sorted(set(matches["team_i"]).union(set(matches["team_j"])))
team_to_id = {t:i for i,t in enumerate(teams)}
n = len(teams)

wins = np.zeros(n)
losses = np.zeros(n)
N = np.zeros((n,n))
```

```python
for _, row in matches.iterrows():
    i = team_to_id[row["team_i"]]
    j = team_to_id[row["team_j"]]

    N[i,j] = 1
    N[j,i] = 1

    if row["result_i"] == "win":
        wins[i] += 1
        losses[j] += 1
    elif row["result_i"] == "loss":
        wins[j] += 1
        losses[i] += 1
```

```python
games = N.sum(axis=1)
print(np.unique(games))

print(pd.Series(wins + losses).value_counts())
```

```python
C = np.zeros((n,n))
b = np.zeros(n)

for i in range(n):
    C[i,i] = 2 + games[i]   # <-- 10 for everyone
    b[i] = 1 + (wins[i] - losses[i]) / 2

for i in range(n):
    for j in range(n):
        if i != j:
            C[i,j] = -N[i,j]

ratings = np.linalg.solve(C, b)
```

```python
colley = pd.DataFrame({
    "team": teams,
    "rating": ratings,
    "wins": wins,
    "losses": losses,
    "games": games
}).sort_values("rating", ascending=False).reset_index(drop=True)

colley.insert(0, "rank", colley.index + 1)

colley.head(36)
```
